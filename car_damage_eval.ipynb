```
import pickle
import cv2,os
import numpy as np
import matplotlib.pyplot as plt
```

#load models

```
model1 = pickle.load(open("/content/drive/My Drive/car-damage-dataset.tar/data1a/model1.pickle", "rb"))
model2 = pickle.load(open("/content/drive/My Drive/car-damage-dataset.tar/data2a/model2.pickle", "rb"))
model3 = pickle.load(open("/content/drive/My Drive/car-damage-dataset.tar/data3a/model3.pickle", "rb"))
```

```
img_arr = cv2.imread(os.path.join('/content/download.jpg'))
new_arr = cv2.resize(img_arr, (224, 224))
image = np.expand_dims(new_arr, axis=0)
image=image/255
```

```
plt.imshow(img_arr)
```

```
damage=['damage','whole']
severity=['minor','moderate','severe']
location=['front','rear','side']
print(damage[np.argmax(model1.predict(image))])
print(location[np.argmax(model2.predict(image))])
print(severity[np.argmax(model3.predict(image))])
```

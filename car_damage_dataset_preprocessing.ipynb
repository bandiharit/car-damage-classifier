#Importing necessary Libraries


```
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import random
import os,cv2
import pickle
```


#Creating data directory and array for classes

```
data_dir= r"/content/gdrive/My Drive/car-damage-dataset.tar/data2a/training"
categories=['00-front','01-rear','02-side']
```

#Defining a function to create training data
* Predefine image_size                      
*For every folder(class) in the given directory, use the cv2 library to grey scale the RGB image and resize it with the predefined img_size.                                                     
*append the image data and corresponding class index to training_data  

```
img_size=224
training_data=[]
def create_training_data():
    for category in categories:
        path = os.path.join(data_dir, category)
        class_num =categories.index(category)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path, img))
                new_arr = cv2.resize(img_arr, (img_size, img_size))
                training_data.append([new_arr, class_num])
            except Exception as e:
                pass

create_training_data()
```

Randomly shuffling training_data to ensure that each data point creates an "independent" change on the model, without being biased by the same points before them.

```
random.shuffle(training_data)
```

#Creating Pickle Files
To save the training_data's serialized format to a file and load it later to directly train the different models without the hassle of data preprocessing again. 

*   Initialize X and Y to store features and labels from training_data respectively.
*   Append features and labels to X and Y
*Similarly X' and Y' as test data


ge

```
X=[]
Y=[]
for features, labels in training_data:
    X.append(features)
    Y.append(labels)

X=np.array(X)
pickle_out=open("featuresrgb.pickle","wb")
pickle.dump(X, pickle_out)
pickle_out.close()
pickle_out=open("labelsrgb.pickle","wb")
pickle.dump(Y, pickle_out)
pickle_out.close()
```
